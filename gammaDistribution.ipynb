```python
import numpy as np
import seaborn as sns
from scipy.stats import gamma, poisson
%matplotlib inline
```

```python
shape, scale = 4., 2. 
s = np.random.gamma(shape, scale, 1000)
```

```python
sns.distplot(s)
```

```python
gamma.fit(s)
```

# Negative Binomial Distribution

```python
from scipy.stats import nbinom
```

```python
r, p = 4, 0.8
nbinorm_data = nbinom.rvs(r, p , size=100)
```

```python
sns.distplot(nbinorm_data)
```

# poisson distribution

```python
p = np.random.poisson(3.87, 2608)
```

```python
sns.distplot(p,kde=False,bins=np.max(p) - np.min(p))
```

#### scipy.statsにはpoisson.fitがないようなので、statsmodelsを使用

```python
import statsmodels.api as sm
```

```python
res = sm.Poisson(p,np.ones_like(p)).fit()
```

```python
res.summary()
```

```python
res.predict()
```
